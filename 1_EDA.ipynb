```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data into a DataFrame
df = pd.read_csv(("movies.csv"))
```

```python
# Data Cleaning and Preparation
# according to what you want to do this part differs greatly
# preprocessing is a big deal when it comes to what you wish your data to reflect and data intergity
```

```python
# Understanding the Structure of the Dataset
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print("\nData types:")
print(df.dtypes)
print("\nSummary statistics:")
print(df.describe())
```

```python
# Visualizing the Distribution of Ratings
plt.figure(figsize=(8, 6))
sns.histplot(df['rating'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Exploring Genre Trends Over the Years
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=df, hue='genre')
plt.title('Count of Movies by Genre Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Genre', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Analyzing the Relationship Between Budget and Gross Earnings
plt.figure(figsize=(8, 6))
sns.scatterplot(x='budget', y='gross', data=df, alpha=0.7)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget (in $)')
plt.ylabel('Gross Earnings (in $)')
plt.show()
```

```python
# Understanding Runtime Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['runtime'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Density')
plt.show()
```
